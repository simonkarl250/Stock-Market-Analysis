```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
dji_file_path = 'DJI.csv'
fchi_file_path = 'FCHI.csv'

# Load data from Excel and CSV
data_dji = pd.read_csv(dji_file_path)  
data_fchi = pd.read_csv(fchi_file_path)

# Display the first few rows of each dataset to understand their structure
print("DJI Data:")
print(data_dji.head())
print("\nFCHI Data:")
print(data_fchi.head())

# Assume both datasets have columns named 'Date' and 'Price'
# Convert 'Date' columns to datetime
data_dji['Date'] = pd.to_datetime(data_dji['Date'])
data_fchi['Date'] = pd.to_datetime(data_fchi['Date'])


data_dji["Close"] = data_dji["Close"].replace(',', '', regex=True).astype(float)
data_dji["Close"]=pd.to_numeric(data_dji["Close"])

# Set 'Date' as index for both DataFrames
data_dji.set_index('Date', inplace=True)
data_fchi.set_index('Date', inplace=True)

# Merge the two DataFrames on 'Date'
data_merged = pd.merge(data_dji, data_fchi, left_index=True, right_index=True, suffixes=('_DJI', '_FCHI'))

# Calculate the correlation between the two stocks

correlation = data_merged['Close_DJI'].corr(data_merged['Close_FCHI'])
print(f'Correlation between DJI and FCHI: {correlation}')

# Plot the historical prices
plt.figure(figsize=(12, 6))
plt.plot(data_merged.index, data_merged['Close_DJI'], label='DJI', color='blue')
plt.plot(data_merged.index, data_merged['Close_FCHI'], label='FCHI', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Historical Prices of DJI and FCHI')
plt.legend()
plt.grid(True)
plt.show()
```
